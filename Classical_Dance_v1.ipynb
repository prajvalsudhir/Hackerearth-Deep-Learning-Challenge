```python
#using pretrained network inceptionv3
```

```python
from tensorflow.keras.applications import InceptionV3,MobileNetV2,resnet50,VGG19,InceptionResNetV2
from tensorflow.keras.models import Sequential,Model
from tensorflow.keras.layers import Input,Conv2D,MaxPool2D,Dense,Dropout,Flatten,BatchNormalization
```

```python
incep_model = InceptionV3(weights="imagenet", include_top=False,
                          input_tensor=Input(shape=(299,299,3)))
```

```python
incep_model.summary()
```

```python
d_model = incep_model.output
d_model = (Flatten())(d_model)
#d_model = MaxPool2D(pool_size=(2,2))(d_model)
#d_model = (Flatten())(d_model)
d_model  = (Dense(1024,activation='relu'))(d_model)
d_model  = (Dropout(0.4))(d_model)
d_model = (BatchNormalization())(d_model)
d_model  = (Dense(512,activation='relu',kernel_regularizer='l2'))(d_model)
d_model  = (Dropout(0.2))(d_model)
# d_model = (BatchNormalization())(d_model)
# d_model  = (Dense(256,activation='relu'))(d_model)
# d_model  = (Dropout(0.1))(d_model)
# d_model = (BatchNormalization())(d_model)
d_model  = (Dense(128,activation='relu'))(d_model)
d_model  = (Dropout(0.1))(d_model)
d_model = (BatchNormalization())(d_model)
d_model  = (Dense(8,activation='softmax'))(d_model)
```

```python
model = Model(inputs = incep_model.input,outputs = d_model)
```

```python
for layer in incep_model.layers:
  layer.trainable = False
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(train_img_gen,validation_data=test_img_gen,epochs=20,batch_size=16)
```

```python
rdf=pd.DataFrame(model.history.history)
rdf
```

```python
rdf[['accuracy','val_accuracy']].plot()
```
